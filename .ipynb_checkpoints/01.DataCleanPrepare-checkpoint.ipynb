### Import Datasets Cab_rides
<hr>

```python
desti = pd.read_csv('cab_rides.csv')
desti.info() #menunjukan data untuk tipe data dan kosong
desti.dropna(inplace= True) #menghapus or drop yang ada na values
desti
```

#### Making Asumsution Distance Depature to Arrive

<hr>

```python
ds = desti['destination'].unique() #tujuan
sr = desti['source'].unique() #jemput
```

```python
awal = [] #Listkosong
tujuan = []
jarak = []
for i in ds:
    for j in sr:
        if i != j:
            a = j
            b = i
            km = round((desti[(desti['destination'] == i) & (desti['source'] == j)]['distance'].mean()),2)
            awal.append(a)
            tujuan.append(b)
            jarak.append(km)
        else:
            break
            
dfJarak = pd.DataFrame()
dfJarak['jemput'] = awal
dfJarak['tujuan'] = tujuan
dfJarak['jarak'] = jarak
dfJarak.dropna(inplace= True)
dfJarak.sort_values('jemput', inplace=True)
dfJarak
```

#### Testing Input Depature to Arrive

<hr>

```python
a = "West End"
b = "Northeastern University"

c = dfJarak[(dfJarak['jemput'] == a) & (dfJarak['tujuan'] == b)]['jarak']
float(c)
```

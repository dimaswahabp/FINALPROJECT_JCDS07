#### Bedasarkan Bar Plot & Diagram Pie : Uber mempunyai banyak customer dibandingkan lyft.

<hr>

```python
sns.set_style('whitegrid')
f, axes = plt.subplots (2,1, figsize=(4,6))

#histogram
x=['Uber','Lyft']
y = [df1.cab_type[(df1.cab_type)=='Uber'].count(),\
     df1.cab_type[(df1.cab_type)=='Lyft'].count()]

vis1= sns.barplot(x,y,palette='Accent',ax=axes[0])
vis1.set(xlabel='Cab Type',ylabel='Number of cab')


for p in vis1.patches:
             vis1.annotate("%.f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='gray', xytext=(0, 20),
                 textcoords='offset points')
#Pie
Mcab_type = [df1.cab_type[(df1.cab_type)=='Uber'].count(),\
             df1.cab_type[(df1.cab_type)=='Lyft'].count()
            ]
     
cab_type = ['Uber','Lyft']

plt.pie(Mcab_type, labels=cab_type, startangle=90, autopct='%.1f%%')


plt.show()
```

### Scatter Plot membandingkan tarif antara Uber dengan Lyft.
<hr>
<p> Sebelum menampilkan grafik harga umum vs jarak, jatuhkan 'Taksi' dari taksi Uber karena mereka menggunakan struktur tarif yang berbeda (dalam dataset Taksi tarif selalu 0 USD) Berdasarkan grafik di atas.<br></p> <p> Dapat kita diamati bahwa Lyft memiliki minimum yang lebih rendah tarif dibandingkan dengan Uber, pada saat yang sama Lyft memiliki tarif maksimum yang lebih tinggi. Juga, jarak perjalanan Uber umumnya lebih tinggi dari Lyft. </p>
<p> Kita juga tahu ada faktor 'pengganda lonjakan tarif' di Lyft, di mana tarif dikalikan dengan tarif tertentu selama jam-jam sibuk.<br></p>
<p> Ketika harga melonjak, Uber tidak menunjukkan pengganda tarif dan sebagai gantinya hanya mengutip harga yang lebih tinggi di muka. Lyft menandai penetapan harga Waktu Perdana dengan persentase: 'Jika tarifnya adalah 50 persen, tarif yang biasanya menjadi USD10 biayanya USD15'. Ini akan diverifikasi dengan visualisasi di bawah:

```python
Taxi=df1[df1.name == 'Taxi'].index
#df1.drop(Taxi, inplace=True)
vis1=sns.scatterplot(x=df1.distance,y=df1.price,data=df1, hue=df1.cab_type, hue_order=cab_type,alpha=0.3, legend='full')
```
